#### Question 1: Find the questions and answers of last month.

```python
# your code here

SITE = StackAPI('stackoverflow')


# questions = SITE.fetch('questions')

question = SITE.fetch('questions', fromdate='2020-09-01', todate='2020-09-30')                         
    
data = pd.DataFrame(question)

data.head()







```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data

```

```python
questions_data = flatten(data, ['items'])

questions_data

questions_data['items_title']

pd.DataFrame(questions_data['items_title'])
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here

questions = SITE.fetch('questions', fromdate='2020-09-01', todate='2020-09-30',
    min=5, tagged='python', sort='votes')


data_2 = pd.DataFrame(questions)

data_2

questions_2 = flatten(data_2, ['items'])

questions_2

questions_2['items_title']


pd.DataFrame(questions_2['items_title']).head()
```
