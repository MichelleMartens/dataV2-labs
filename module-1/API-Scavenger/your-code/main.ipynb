## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
# your code here

hidden_joke = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger', auth=(username,token))

# cold_joke = pd.DataFrame(hidden_joke.json())

# cold_joke

find_files = requests.get('https://api.github.com/search/code?q=filename:.scavengerhunt+repo:ironhack-datalabs/scavenger', auth=(username,token))

find = find_files.json()

pd.DataFrame(find).head()

# .0001.scavengerhunt to .0024.scavengerhunt.
`````

#### 2.  Sort the filenames ascendingly.

`````python
# your code here

# sort = requests.get('https://api.github.com/search/code?q=filename:.scavengerhunt+repo:ironhack-datalabs/scavenger&sort&order=asc', auth=(username,token))
# sort_1 = requests.get('https://api.github.com/search/code?q=filename:.scavengerhunt+repo:ironhack-datalabs/scavenger&sort=name&direction=asc', auth=(username,token))

sort = requests.get('https://api.github.com/search/code?q=filename:.scavengerhunt+repo:ironhack-datalabs/scavenger', auth=(username,token))


sort = sorted(sort.json()['items'], key=lambda x: x['name'])

sort[0]

`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
# your code here

lst = []

headers = {'Accept': 'application/vnd.github.v3.raw'}

for x in sort: 
    content_path = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{}'.format(x['path'])
    response = requests.get(content_path, headers=headers, auth=(username, token))
    lst.append(response.text)
    
lst


# headers = {'Accept': 'application/vnd.github.v3.raw'}

`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
# your code here

content = [word.strip('\n') for word in lst]
content

joke = ' '.join(content)

joke
`````

#### 5. Print out the joke.

`````python
# your code here

print(joke)
`````
